```python
def wrap(string, max_width):
    rows = []
    i =0
    
    while i < len(string):
        row = string[i:max_width+i]
        rows.append(row)
        i += max_width
    strings = ""
    for j in rows:
        strings = strings + j +"\n"
    return strings
    
```

```python
wrap( 'ABCDEFGHIJKLIMNOQRSTUVWXYZ',4)
```

```python
#Using textWrap
import textwrap

def wrap(string, max_width):
    return textwrap.fill(string,max_width)
```

```python
wrap( 'ABCDEFGHIJKLIMNOQRSTUVWXYZ',4)
```
