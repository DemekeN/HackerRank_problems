```python
if __name__ == '__main__':
    python_students = []
    for _ in range(int(input())):
        name = input()
        score = float(input())
        python_students.append([name,score])
    second_lowest = sorted(list(set([x[1] for x in python_students])))[1]
    selected_students = []
    for i in python_students:
        if i[1]== second_lowest:
            selected_students.append(i[0])
    print("\n".join(sorted(selected_students)))
```
