```python
N = int(input())

items_list=[]

for _ in range(N):
    line = input().split()
    comand,element = line[0], line[1:]
    nos = list(map(int, element))
    if comand=='insert':
        items_list.insert(nos[0],nos[1])
    elif comand=='reverse':
        items_list.reverse()
    elif comand == 'append':
        items_list.append(nos[0])
    elif comand=='remove':
        items_list.remove(nos[0])
    elif comand=='print':
        print(items_list)
    elif comand=='pop':
        items_list.pop()
    elif comand=='sort':
        items_list.sort()
    
```
