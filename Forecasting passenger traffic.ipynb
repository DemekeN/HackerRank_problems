```python
n, N = 12, int(input())
Y = [int(input().split()[1]) for i in range(N)]
m = float(max(Y))
Y = [i / m for i in Y]
X = [[i] + [1 if j == i % n else 0 for j in range(n)] for i in range(N)]


from sklearn.ensemble import GradientBoostingRegressor
params = {'n_estimators': 800, 'max_depth': 8, 'min_samples_split': 2,
          'learning_rate': 0.01, 'loss': 'ls'}
clf = GradientBoostingRegressor(**params)
clf.fit(X, Y)


X = [[N + i] + [1 if j == (N + i) % n else 0 for j in range(n)] for i in range(12)]
for i in clf.predict(X):
    print(i * m)
```
