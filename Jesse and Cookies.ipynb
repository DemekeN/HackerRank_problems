```python
from heapq import heappush, heappop
def cookies(k, A):
    packed = []
    rt = 0

    for elt in A:
        heappush(packed, elt)

    while any(c < k for c in packed) and len(packed) > 1:
        last = heappop(packed)
        prelast = heappop(packed)

        new = last + 2*prelast

        heappush(packed, new)
        rt += 1

    if all(c >= k for c in packed):
        print(rt)
    else:
        print(-1)
    
n, k = map(int, input().strip().split(' '))
A = list(map(int, input().strip().split(' ')))
cookies(k, A)
```
